```python
import pandas as pd
from sklearn.linear_model import Perceptron

# Data Loading
path = "/content/Social_Network_Ads.csv"
df = pd.read_csv(path, sep=",")

# Data Preprocessing
df['Male'] = 0
df['Female'] = 0
df.loc[df['Gender'] == 'Male', 'Male'] = 1
df.loc[df['Gender'] == 'Female', 'Female'] = 1
df.drop(['Gender', 'User ID'], axis=1, inplace=True)
df = df[["Age", "Male", "Female", "EstimatedSalary", "Purchased"]]

# Data Splitting
dfxtrain = df.iloc[:319, :-1]  # Exclude the "Purchased" column
dfxtest = df.iloc[320:, :-1]   # Exclude the "Purchased" column

# Target Variable and Output
y = df["Purchased"]
dfytrain = y.iloc[:319]
dfytest = y.iloc[320:]

# Model Training and Evaluation
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(dfxtrain, dfytrain)

# Model Prediction
predictions = clf.predict(dfxtest)

# Output Printing
print("Ground Truth:")
print(dfytest)
print("Predictions:")
print(predictions)
```
